```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
%precision 4
plt.style.use('ggplot')
```

```python
def f1(n,k):
  return n*n

def f2(n,k):
  return np.log(n)*k*n

n=np.arange(0,20001)

plt.plot(n, f1(n,1),c='blue')
plt.plot(n,f2(n,1000),c='red')
plt.xlabel('Size of input (n)',fontsize=16)
plt.ylabel('Number of operations',fontsize=16)
plt.legend(['$\mathcal{0}(n^2)$','$\mathcal{0}(n \log n)$'],loc='best',fontsize=20)
```
